```python
import kagglehub
import os
import matplotlib.pyplot as plt
import numpy as np
# Download latest version
path = kagglehub.dataset_download("tongpython/cat-and-dog")
```

```python
images_path = os.path.join(path, "training_set/training_set")

# Explore each category and display n sample images 
n = 10
for category in os.listdir(images_path):
    category_path = os.path.join(images_path, category)
    num_images = len(os.listdir(category_path))
    print(f"Number of images in category '{category}': {num_images}")
    # Show n sample images
    sample_images = np.random.choice(os.listdir(category_path), n, replace=False)
    plt.figure(figsize=(10, 5))
    for i, img_name in enumerate(sample_images):
        img_path = os.path.join(category_path, img_name)
        img = plt.imread(img_path)
        plt.subplot(1, n, i + 1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"{category} {i+1}")
    plt.show()
```
